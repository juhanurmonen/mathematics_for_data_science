# Multivariate calculus

We earlier considered examples of functions with one variable, and also vectors with several dimensions.

In **multivariate** (or **multivariable**) **calculus** one variable case is extended to the one where functions can have several variables. This concerns both the domain and the range of the function.

<u>Formally</u>, a multivariate function is function $f:\mathbf{R}^m\to\mathbf{R}^n$, where $m\geq 2$, $n\geq 1$. Here $m$-dimensional vectors are mapped to $n$-dimensional vectors. Such functions are often also called *vector-valued* functions. The special case, where $m=n=3$ is traditionally called *vector analysis*.

Therefore, a multivariate functions depends on several input variables to produce an output.

<u>Some general examples</u> of multivariate functions include:
*  **Surfaces**: Here $f:\mathbf{R}^2\to\mathbf{R}^n$, and $n>2$. Areas of surfaces are analysed. <img title = "Paraboloid" alt = "image of a paraboloid" width=80 height = 30 src="https://upload.wikimedia.org/wikipedia/commons/thumb/1/12/Paraboloid_of_Revolution.svg/1024px-Paraboloid_of_Revolution.svg.png">
* **Scalar fields**: Here $f:\mathbf{R}^n\to\mathbf{R}$. Finding local maxima and minima and directional derivatives (the rate at which a function changes in a particular direction) are analysed. <a href="https://en.wikipedia.org/wiki/Scalar_field"><img alt = "image of a scalar field" width=80 height=40 src="https://upload.wikimedia.org/wikipedia/commons/thumb/a/a8/Scalar_field.png/220px-Scalar_field.png"></a>

* **Vector fields**: Here $f:\mathbf{R}^m\to\mathbf{R}^n$. Gradients and divergences over a vector space are analysed. <a href="https://en.wikipedia.org/wiki/Vector_field"><img alt = "image of a portion of a vector field" width=80 height=40 src="https://upload.wikimedia.org/wikipedia/commons/thumb/b/b9/VectorField.svg/250px-VectorField.svg.png"></a>

### Limits

Limits are many ways more complicated in case of multivariate case compared to univariate.


Consider a parametrised path $s(t):\mathbf{R}\to\mathbf{R}^m$ in $m$-dimensional space. Any function $f(\vec{x}):\mathbf{R}^m\to\mathbf{R}^n$ can be projected on the path $s$ as a function $f(s(t)):\mathbf{R}\to\mathbf{R}^n$.

The limit of $f$ to the point $s(t_0)$ along the path $s(t)$ is now defined as $$\lim_{\vec{x}\to s(t_0)} f(\vec{x}) = \lim_{t\to t_0} f(s(t)).$$

A general limit can be defined if the limits to a point along all possible paths converge to the same value. That is, the limit of $f:\mathbf{R}^m\to\mathbf{R}^n$ to some point $x_0\in\mathbf{R}^m$ is defined to be $a$, if and only if, $$\lim_{t\to t_0} f(s(t)) = a$$ _for all continuous functions_ $s(t):\mathbf{R}\to\mathbf{R}^m$ for which $s(t_0)=x_0$.

It is important to notice that the limit does does not only depend on the point which the limit approaches. The limit may also depend on the form of $s(t)$ &mdash; the path chosen may be relevant.

We illustrate this property by considering the function $$f(x,y) =  \frac{x^2\cdot y}{x^4+y^2}$$ and the limit at the point $(0,0)$ by approaching it
* along the line $y=k\cdot x$ and
* along the path $y=\pm x^2$.

```python
# Plot of the functions f(x,y) = (x^2 y)/(x^4 + y^2)

limit_example = (x**2 * y) / (x**4 + y**2)

plot3d(limit_example, (x, -2, 2), (y, -4, 4))
```

The line $y=k\cdot x$ is in parametric form $x(t)=t$, $y(t)=k\cdot t$ and the limit above along this line will be $$\lim_{t\to 0}f(x(t), y(t)) = \lim _{t\to 0}\frac{k\cdot t^3}{t^4+k^2\cdot t^2}=0.$$

The path $y=\pm x^2$, in turn, is in parametric form $x(t)=t$, $y(t)=\pm t^2$ and the limit along this path comes to $$\lim_{t\to 0}f(x(t),y(t)) = \lim _{t\to 0}\frac{\pm t^4}{t^4 + t^4} = \pm\frac{1}{2}.$$

Hence a general limit at the point $(0,0)$ cannot be defined for this function.

### Continuity briefly

Continuity also is many ways more complicated in case of multivariate case compared to univariate. Following the same line of thinking as for the limit value, continuity can be defined as follows.

A multivariate function $f:\mathbf{R}^m\to\mathbf{R}^n$ is _continuous_ at the point $x_0\in\mathbf{R}^m$, if and only if, $$\lim_{t\to t_0} f(s(t)) = f(x_0)$$ _for all continuous functions_ $s(t):\mathbf{R}\to\mathbf{R}^m$ for which $s(t_0)=x_0$.

Notice that being continuous along one path $s(t)$ does not imply multivariate continuity.

In particular, a function $f(x,y)$ defined on $\mathbf{R}^2$, may be continous on both coordinates $x$ and $y$ even though the function $f$ is not is multivariate continuous.

### Properties of multivariate limits and continuity

* _Linearity_: For all multivariate continuous functions $f,g:\mathbf{R}^m\to\mathbf{R}^n$ and real numbers $\alpha,\beta\in\mathbf{R}$, $$\alpha\cdot f(\vec{x}) + \beta\cdot g(\vec{x})$$ is also a multivariate continuous  function
* _Composition_: If $f:\mathbf{R}^m\to\mathbf{R}^n$ and $g:\mathbf{R}^n\to\mathbf{R}^p$ are multivariate continuous functions at the points $\vec{x}\in\mathbf{R}^m$ and $f(\vec{x})\in\mathbf{R}^n$, then $(g\circ f):\mathbf{R}^m\to\mathbf{R}^p$ is a multivariate continuous function at the point $\vec{x}$
* _Multiplication_: If $f,g:\mathbf{R}^m\to\mathbf{R}$ are multivariate continous functions at the point $\vec{x}\in\mathbf{R}^m$, then $(f\cdot g):\mathbf{R}\to\mathbf{R}$ is multivariate continous function at the point $\vec{x}$.
* _Quotient_: If $f,g:\mathbf{R}^m\to\mathbf{R}$ are multivariate continous functions at the point $\vec{x}\in\mathbf{R}^m$, then $(f/g):\mathbf{R}\to\mathbf{R}$ is multivariate continous function at the point $\vec{x}$ assuming that $g(\vec{x})\not= 0$.
* If $f:\mathbf{R}^m\to\mathbf{R}$ is a multivariate continuous function at the point $\vec{x}\in\mathbf{R}^m$, then $|f|$ is a multivariate continous function at the point  $\vec{x}$.

## Differentiation of Multi-Variate Functions

When we consider univariate (one variable) functions, finding the derivative of the function is often essential. In that way we can study the rate of change.


## Partial Derivative

The concept of differentiation &ndash; finding the derivative &ndash; can be generalized to multivariate functions as well. This generalisation is called the **gradient**.

The technique behind finding the gradient of a multivariate function is that we go through all the variables in order to find the derivative of the function with respect to that variable, and considering all the other variables as constant during the calculation. The **gradient** is then the collection of all these **partial derivatives**.  

Consider, as an example, the function $f$ introduced and defined above. We first calculate the partial derivative with respect to the variable $x$ and considering $y$ as a constant. That is $$\frac{\delta f(x,y)}{\delta x}=\frac{\delta}{\delta x}f(x,y)= \frac{\delta}{\delta x}(3\cdot x^2+2\cdot y^2)=6\cdot x.$$

```python
# Above calculation in SymPy

f.diff(x)
```

Intuitition behind this is that when $y=0$ there is a plane passing through the parabolic surface. The resulting cross-section of $f(x,y)$ is the quadratic curve $3\cdot x^2$. We calculate the derivative (steepness, slope) of the parabolic surface in the direction of $x$.

We similarly find the cross-section of the parabolic surface by slicing it with a plane passing through a constant value of the variable $x$. Technically, we keep $x$ as constant and find the partial derivative of $f(x,y)$ with respect  to the variable $y$:

$$\frac{\delta f(x,y)}{\delta y}=\frac{\delta}{\delta y}f(x,y)= \frac{\delta}{\delta y}(3\cdot x^2+2\cdot y^2)=4\cdot y.$$

```python
# Above calculation in SymPy

f.diff(y)
```

Now we have found the univariate derivative of $f(x,y)$ in both directions $x$ and $y$. Combining the results gives us the gradient of $f(x,y)$:

$$\frac{d f(x,y)}{d(x,y)}=[\frac{\delta f(x,y)}{\delta x}, \frac{\delta f(x,y)}{\delta y}] = [6\cdot x, 4\cdot y].$$

```python
# Above calculation in SymPy

[f.diff(var) for var in [x,y]]
```

### Multiple Integration briefly

The concept of multiple integral extends the univariate integral case to the any number of variables. Double and triple integrals can be used to calculate areas of regions in plane and volumes of regions in space.

A multiple integral can be calculated provided that the integrand function is continuous throughout the domain of integration.

In particular, the followimg holds.

Suppose $X\subseteq\mathbf{R}$ and $Y\subseteq\mathbf{R}$ and $f(x,y)$ is continuous function on the rectangle $X\times Y\subseteq\mathbf{R}^2$. Then the double integral can be evaluated as an iterated integral:
$$\int\int_{X\times Y}f(x,y)d(x,y)=\int_{X} \left(\int_{Y} f(x,y)dy \right) dx = \int_{Y} \left(\int_{X} f(x,y) dx\right) dy.$$

### Multiple integrals in SymPy

Integration with multiple variables with SymPy can be done following the pattern we learned from differentiation.

We integrate $$\int e^{-(x+y)}dxdy.$$

```python
from sympy import integrate, exp, oo
integrate(exp(-(x+y)), x, y)
```

Compare this with the next one.

```python
integrate(integrate(exp(-(x+y)), x), y)
```

Determine $$\int_{-\infty}^{\infty}\int_{-\infty}^{\infty}e^{-x^2-y^2}dxdy.$$

```python
expr_e_xy_infty = integrate(exp(-x**2-y**2), (x, -oo, oo), (y, -oo, oo))
expr_e_xy_infty
```

Consider the function $f(x, y, z) = 1$ over the unit cube and calculate $$\int_0^1\int_0^1\int_0^1 f(x,y,z)dxdydz.$$

```python
f_unit_cube = 1
integrate(f_unit_cube, (x, 0, 1), (y, 0, 1), (z, 0, 1))
```

Above, we verified that the volume of the unit cube is one, as could be expected. We can also do more.

Consider the following density function $f_\mathrm{density}:\mathbf{R}^3\to\mathbf{R}$ such that $$f(x, y, z) = x + y + z.$$ Now the volume integral will give us the total mass of the cube.

```python
f_xyz_density = x + y + z
integrate(f_xyz_density, (x, 0, 1), (y, 0, 1), (z, 0, 1))
```

### Applications

Multivariate calculus has numerous application areas. To mention a couple of them, in _regression analysis_ it is used to derive formulas for estimating relationships among various sets. In context with _neural networks_ multivariate calculus is used diversely to update model parameters or weights.

Neural network contains many weights and the error of each weight is a function. In minimising error functions, an optimisation algorithm is trying to look for its gradient downhill. That's why updating the weight values requires that the gradient of the error function is calculated with respect to all of these weights.

The gradient of the error function is calculated by finding the partial derivative of the error with respect to each weight. Here each derivative of the error function is calculated keeping other weights constant. This way weights are updated independently and an optimal set of weights is found.
